```
#file is used to do some deeper analysis of the data based on daily basis & effect on bicycle count due weekends and other holidays
#importing necessary lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
#read file with proper date formatting
DF = pd.read_csv("daily_data.csv", parse_dates = ['Date'], index_col = ['Date'])
```

```
#compare yearly traffic on huron st
fig, ax = plt.subplots(figsize = (7,7))

#plot the graph
ax.plot(DF.index.values, DF['TH'], color = 'red')
ax.plot(DF.index.values, DF['EBH'], color = 'blue')
ax.plot(DF.index.values, DF['WBH'], color = 'green')

plt.setp(ax.get_xticklabels(), rotation=45)
plt.grid(True)

ax.set(xlabel="Date", ylabel="Total Traffic", title="Yearly Traffic on Huron Street");
```

```
#compare yearly traffic on markham st
fig, ax = plt.subplots(figsize = (7,7))
#plot
ax.plot(DF.index.values, DF['TM'], color = 'red')
ax.plot(DF.index.values, DF['EBM'], color = 'blue')
ax.plot(DF.index.values, DF['WBM'], color = 'green')

plt.setp(ax.get_xticklabels(), rotation=45)
plt.grid(True)

ax.set(xlabel="Date", ylabel="Total Traffic", title="Yearly Traffic on Markham Street");
```
